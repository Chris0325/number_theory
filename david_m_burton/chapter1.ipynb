```python
import math
import numpy as np
```

```python
# 1.1 5
for n in range(8):
    m = math.factorial(n) + 1
    if np.sqrt(m).is_integer():
        print(n)
```

```python
# 1.1 12
def t(n):
    if n % 2:
        return (3 * n + 1) // 2
    else:
        return n // 2


n = 21
while n != 1:
    print(n, end=' ')
    n = t(n)

print()  
n = 23
while n != 1:
    print(n, end=' ')
    n = t(n)
```
